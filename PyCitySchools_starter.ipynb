```python
# Dependencies and Setup
import pandas as pd
import numpy as np
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
Total_schools = len(school_data_complete["school_name"].unique())
Total_students = (school_data_complete["student_name"].count())
Total_budget = school_data_complete["budget"].unique().sum()
Avg_math_score = school_data_complete["math_score"].mean()
Avg_reading_score = school_data_complete["reading_score"].mean()
Pass_in_math = len(school_data_complete[school_data_complete["math_score"] >= 70])
Pass_in_reading = len(school_data_complete[school_data_complete["reading_score"] >= 70])
Pass_in_both = len(school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)])

results = pd.DataFrame({
    "Total Schools" :[Total_schools] ,
    "Total Students" : f'{Total_students:,}',
    "Total Budget" : f'${Total_budget:,}' ,
    "Average Math Score" : Avg_math_score ,
    "Average Reading Score" : Avg_reading_score,
    "% Passing Math" : (Pass_in_math/Total_students) * 100,
    "% Passing Reading" : (Pass_in_reading/Total_students) * 100,
    "% Overall Passing" :(Pass_in_both/Total_students) * 100
})
results
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_group = school_data_complete.groupby("school_name")
school_type = pd.DataFrame(school_group["type"].unique())

Total_student = pd.Series(school_group["student_name"].count())
Total_school_budget = pd.Series(school_group["budget"].unique())
per_student_budget = pd.Series(Total_school_budget/Total_student)
avg_math_score = pd.Series(school_group["math_score"].mean())
avg_reading_score = pd.Series(school_group["reading_score"].mean())

df_student_passing_math = school_data_complete[school_data_complete["math_score"] >= 70]
group_passing_math = df_student_passing_math.groupby("school_name")
percentage_passing_math = pd.DataFrame((group_passing_math["student_name"].count()/Total_student) *100)


df_student_passing_reading = school_data_complete[school_data_complete["reading_score"] >= 70]
group_passing_reading = df_student_passing_reading.groupby("school_name")
percentage_passing_reading = pd.DataFrame((group_passing_reading["student_name"].count()/Total_student) *100)


df_student_passing_overall = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]
group_passing_overall = df_student_passing_overall.groupby("school_name")
percentage_passing_overall = pd.DataFrame((group_passing_overall["student_name"].count()/Total_student) *100)

school_summary = pd.DataFrame({
    "School Type" : school_type["type"],
    "Total Students" :Total_student,
    "Total School Budget" : Total_school_budget,
    "Per Student Budget" :per_student_budget,
    "Average Math Score" : avg_math_score,
    "Average Reading Score" :avg_reading_score,
    "% Passing Math" :percentage_passing_math["student_name"] ,
    "% Passing Reading" :percentage_passing_reading["student_name"],
    "% Passing Overall" :percentage_passing_overall["student_name"]
})
school_summary
#school_type
```

## Math Scores by Grade

```python
grade_9th = school_data_complete.loc[school_data_complete["grade"] == '9th',["school_name","math_score"]]
grade_9th_group = grade_9th.groupby("school_name")
grade_9th_avg_math_score = pd.Series(grade_9th_group["math_score"].mean())

grade_10th = school_data_complete.loc[school_data_complete["grade"] == '10th',["school_name","math_score"]]
grade_10th_group = grade_10th.groupby("school_name")
grade_10th_avg_math_score = pd.Series(grade_10th_group["math_score"].mean())

grade_11th = school_data_complete.loc[school_data_complete["grade"] == '11th',["school_name","math_score"]]
grade_11th_group = grade_11th.groupby("school_name")
grade_11th_avg_math_score = pd.Series(grade_11th_group["math_score"].mean())

grade_12th = school_data_complete.loc[school_data_complete["grade"] == '12th',["school_name","math_score"]]
grade_12th_group = grade_12th.groupby("school_name")
grade_12th_avg_math_score = pd.Series(grade_12th_group["math_score"].mean())

math_avg_scores_by_grade= pd.DataFrame({
    "9th" : grade_9th_avg_math_score,
    "10th" :grade_10th_avg_math_score,
    "11th" :grade_11th_avg_math_score,
    "12th" :grade_12th_avg_math_score
    
})
math_avg_scores_by_grade
```
